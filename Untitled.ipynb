```python
reg102 = [x for x in df.columns if str(x)[-1:] !="_"]
time_range = list(map(lambda x:x[:13],df.index.astype(str)))
```

```python
def df_plot(region=1075, start='2016-05-31 23', finish='2016-06-30 17'):
    if datetime(int(start[:4]),int(start[5:7]),int(start[8:10]),int(start[11:])) > datetime(int(finish[:4]),int(finish[5:7]),int(finish[8:10]),int(finish[11:])):
        finish = '2016-05-31 00'
    #print (region)
    region = int(region)
    df.loc[:,[region,str(region)+'_']][start:finish].rename(columns={region: "fact", str(region)+'_': "prediction"}).plot.line(title='NY yellow taxi pick-ups for region '+str(region), figsize=(20, 8))
```

```python
import folium
import branca.colormap as cmap

points = regions.loc[reg102,:]

# Границы карты
west, south, east, north = -74.25559, 40.49612, -73.70001, 40.91553
```

```python
url_base = 'http://server.arcgisonline.com/ArcGIS/rest/services/'
service = 'NatGeo_World_Map/MapServer/tile/{z}/{y}/{x}'
tileset = url_base + service

im = folium.Map(location=[(north+south)/2, (east+west)/2], zoom_start=11,
                     control_scale = True, tiles=tileset, attr='USGS style')

def make_map(timepoint):
    j=-1
    
    # Создадим вектор проголарифмирвоанных и отнормирванных значений количества поездок в среднем за месяц, для определения цветов
    norm = matplotlib.pyplot.Normalize()
    a = df.loc[timepoint,reg102].values
    a[a < 0] = 0
    b = np.log(a)
    b[b < 0] = 0
    colors_1= norm(b)
    #colors_1 = norm(np.log(df.loc[timepoint,regions_ids]))
    linear = cmap.LinearColormap(['#2166ac', 'red'], index=[min(colors_1), max(colors_1)])
    
    for i in points.index:
        j += 1
        rgb = linear.rgb_hex_str(colors_1[j])
        folium.features.RectangleMarker(bounds=[[points.ix[i].south, points.ix[i].west], [points.ix[i].north, points.ix[i].east]],
                                        color = 'transparent',
                                        fill_color = rgb,
                                        fill_opacity = 0.5,
                                        popup = "Region: " + str(int(points.ix[i].region)) + \
                                                "\n Mean pickups per hour: " + str(round(df.loc[timepoint,int(points.ix[i].region)],1))
                                       ).add_to(im)

    colormap = cmap.LinearColormap(['#2166ac', 'red']).scale(1, int(max(df.loc[timepoint,reg102].values)))
    colormap.caption = 'A colormap caption'
    im.add_child(colormap)
    return im
```
